```python
import time
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize the Gaussian Naive Bayes classifier
clf = GaussianNB()

# Assuming x_train, x_test, y_train, y_test are predefined and split properly

# Calculate the training time
start_train_time = time.time()
clf.fit(x_train, y_train)
end_train_time = time.time()

# Calculate the prediction time
start_pred_time = time.time()
y_pred = clf.predict(x_test)
end_pred_time = time.time()

# Calculate training and prediction times
training_time = end_train_time - start_train_time
prediction_time = end_pred_time - start_pred_time

# Calculate accuracy
g_accuracy = accuracy_score(y_test, y_pred)

# Calculate error rate
error_rate = 1 - g_accuracy

# Output the results
print("Accuracy:", g_accuracy)
print(f"Error Rate: {error_rate * 100:.2f}%")
print(f"Training Time: {training_time:.4f} seconds")
print(f"Prediction Time: {prediction_time:.4f} seconds")

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Print confusion matrix
print("\nConfusion Matrix:")
print(cm)

# Plot confusion matrix (optional, for visualization)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
